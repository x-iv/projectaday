```python
import requests

def convert_json_to_txt(json_url, txt_filename):
    response = requests.get(json_url)
    data = response.json()

    with open(txt_filename, 'w') as file:
        for item in data['words']:
            target_word = item['targetWord']
            english_word = item['englishWord']
            file.write(f"{target_word}: {english_word}\n")

    print("JSON file converted to TXT successfully.")

# Specify the URL of the JSON file on GitHub
json_url = 'https://raw.githubusercontent.com/SMenigat/thousand-most-common-words/master/words/fr.json'
# Specify the desired filename for the resulting text file
txt_filename = 'frenchtoenglish.txt'

# Convert the JSON file to TXT
convert_json_to_txt(json_url, txt_filename)
```

```python
#hangman using french
import random
import string

def select_random_word():
    with open('jusfrench.txt', 'r') as file:
        word_bank = file.read().splitlines()
        return random.choice(word_bank)

def get_user_guess():
    while True:
        guess = input("Enter your guess (a single letter): ").lower()
        if len(guess) == 1 and guess.isalpha():
            return guess
        else:
            print("Invalid input. Please enter a single letter.")

def update_word_display(word, guessed_letters):
    displayed_word = ""
    for letter in word:
        if letter in guessed_letters:
            displayed_word += letter
        else:
            displayed_word += "_"
    return displayed_word

def play_hangman():
    max_attempts = 6
    guessed_letters = []
    word = select_random_word()
    attempts = 0

    while attempts < max_attempts:
        print("\nWord:", update_word_display(word, guessed_letters))
        print("Attempts left:", max_attempts - attempts)
        print("Guessed letters:", ", ".join(guessed_letters))

        guess = get_user_guess()

        if guess in guessed_letters:
            print("You already guessed that letter.")
        elif guess in word:
            print("Correct guess!")
            guessed_letters.append(guess)
            if word == update_word_display(word, guessed_letters):
                print("\nCongratulations! You guessed the word:", word)
                break
        else:
            print("Incorrect guess.")
            guessed_letters.append(guess)
            attempts += 1

    if attempts == max_attempts:
        print("\nGame over! You ran out of attempts. The word was:", word)

    play_again = input("\nDo you want to play again? (yes/no): ")
    if play_again.lower() == "yes":
        play_hangman()
    else:
        print("Thank you for playing Hangman!")

# Start the game
print("Welcome to Hangman!")
play_hangman()
```
